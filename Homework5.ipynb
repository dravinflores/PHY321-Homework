* In 3.3 of Taylor, we are given the following formulas:

$$
    \tag{1}
    \mathbf{R} = \frac{1}{M} \sum_{\alpha=1}^{N} m_{\alpha} \mathbf{r}_{\alpha}
$$

$$
    \tag{2}
    M = \sum_{\alpha=1}^{N} m_{\alpha}
$$

* We are also given two bodies, with masses $M_1$ and $M_2$ and center of masses $\mathbf{R}_1$ and $\mathbf{R}_2$. 

* We will create an inertial frame of reference with the common origin $O$, so that we can find the center of mass for this system.

* We should note that:

$$
    M_1 = \sum_{\alpha=1}^{N} m_{1, \alpha}, \quad M_2 = \sum_{\alpha=1}^{N} m_{2, \alpha}
$$

* We should also note that

$$
    \mathbf{R}_1 = \frac{1}{M_1} \sum_{\alpha=1}^{N} m_{1, \alpha} \ \mathbf{r}_{1, \alpha}, \quad \mathbf{R}_2 = \frac{1}{M_2} \sum_{\alpha=1}^{N} m_{2, \alpha} \ \mathbf{r}_{2, \alpha}
$$

* We know that the net mass can be given using formula (1) as:

$$
    M' = \sum_{\alpha=1} M_{\alpha} = M_1 + M_2 = \sum_{\alpha=1}^{N} m_{1, \alpha} + \sum_{\alpha=1}^{N} m_{2, \alpha} = \sum_{\alpha=1}^{N} m_{1, \alpha} + m_{2, \alpha}
$$

* We also know that we can find the center of mass using formula (2) as:

$$
    \mathbf{R}' = \frac{1}{M'} \sum_{\alpha=1}^{N} M_{\alpha} \mathbf{R}_{\alpha} = \frac{1}{M_1} \sum_{\alpha=1}^{N} m_{1, \alpha} \ \mathbf{r}_{1, \alpha} + \frac{1}{M_2} \sum_{\alpha=1}^{N} m_{2, \alpha} \ \mathbf{r}_{2, \alpha} = \frac{1}{M'} \left( M_{1} \mathbf{R}_{1} + M_{2} \mathbf{R}_{2} \right) = \frac{M_{1} \mathbf{R}_{1} + M_{2} \mathbf{R}_{2}}{M_1 + M_2}
$$

* This result follows from the two equations above, as well as from the idea layed out in Chapter 3 of Taylor, that a center of mass can be treated as a point particle with a particular mass $M$ and a position $\mathbf{R}$ as the net force of this particle can be found as

$$
    \mathbf{F}^{\text{ext}} = M \ddot{\mathbf{R}}
$$

* The equation above helps us to understand why our result is valid. Thus, we can treat the center of mass as a good representation of the position of the system, the sum of the mass as a good representation of the mass of the system, and package everything into a single particle, which then allows us to talk about that system's motion simply.

6A.

* We are given the function

$$
    \frac{d^2 x}{dt^2} = -\omega_{0}^{2} x
$$

* We are also given the initial conditions: 

$$
    x(0) = 1, \quad x'(0) = 0
$$

* We can find the solution to this differential equation using the initial conditions as:

$$
    x(t) = cos(\omega t)
$$

```python
# Exercise 6

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import math
%matplotlib inline

def calculate_analytic(tf=10.0, ti=0.0, dt=0.001, omega=1, r0=0):
    n = math.ceil( (tf - ti) / dt)
    t = np.zeros(n)
    x = np.zeros(n)
    v = np.zeros(n)
    x[0] = r0
    
    for i in range(n - 1):
        t[i+1] = t[i] + dt
        x[i+1] = math.cos(omega * t[i])
        
    return (t, x)
        

def get_acceleration(g=0, v=0, x=0):
    # Here we expect gamma to be zero. 
    return -2*g*v - x

def calculate_euler(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros(n)
    v = np.zeros(n)

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0

    for i in range(n - 1):
        a = get_acceleration(0, 0, r[i])
        
        v[i+1] = v[i] + a*dt
        r[i+1] = r[i] + v[i]*dt
        t[i+1] = t[i] + dt
        
    return (t, r, v)


def calculate_euler_cromer(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros(n)
    v = np.zeros(n)

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        a = get_acceleration(0, 0, r[i])
        
        v[i+1] = v[i] + a*dt
        r[i+1] = r[i] + v[i+1]*dt
        t[i+1] = t[i] + dt
        
    return (t, r, v)


def calculate_velocity_verlet(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros(n)
    v = np.zeros(n)

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        a = get_acceleration(0, 0, r[i])
        
        r[i+1] = r[i] + v[i]*dt + 0.5*a * (dt ** 2)
        ap = get_acceleration(0, 0, r[i+1])
        
        v[i+1] = v[i] + 0.5*(a + ap)*dt
        t[i+1] = t[i] + dt
        
    return (t, r, v)


# Intial Conditions
r_0 = 1
v_0 = 0

# calculate_velocity_verlet(1, 0.001, 10, 0, r0, v0)

# (ta, ra, va) = calculate_euler(r0=r_0)
(ta, ra) = calculate_analytic(omega=1, r0=r_0)
(tb, rb, vb) = calculate_euler_cromer(r0=r_0)
(tc, rc, vc) = calculate_velocity_verlet(r0=r_0)

# Plot the data
# (figure, axis) = plt.subplots()
# axis.set_xlabel('x (m)')
# axis.set_ylabel('y (m)')
# axis.plot(ra[:, 0], ra[:, 1])
# figure.tight_layout()


plt.subplot(1, 3, 1)
plt.plot(ta, ra)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title("Analytic")
# plt.tight_layout()

plt.subplot(1, 3, 2)
plt.plot(tb, rb)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title("Euler Cromer's Method")
# plt.tight_layout()

plt.subplot(1, 3, 3)
plt.plot(tc, rc)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title("Velocity Verlet's Method")
# plt.tight_layout()

plt.show()
```
