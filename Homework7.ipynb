> ### **A**.

We know that there are only two forces, and that they are an action-reaction pair. This means that the following is true:

$$
    \sum_{i \ne j} \mathbf{F}_{ij} = 0
$$

This means that the sum of all internal forces is zero. This would mean that the following must hold:

$$
    m \ \mathbf{a} = 0 \quad \Rightarrow \quad \sum_{i} \frac{d}{dt} \ \mathbf{p}_i = 0
$$

If we consider the center of mass position $\mathbf{R}$, which is given as:

$$
    \mathbf{R} = \frac{m_1 \mathbf{r}_1 + m_2 \mathbf{r}_2}{M}, \ M = m_1 + m_2
$$

Then we can consider the total momentum as:

$$
    \mathbf{P} = M \dot{\mathbf{R}}
$$

Seeing as the sum of the internal forces is zero, this must mean that the following is also true:

$$
    \frac{d}{dt} \mathbf{P} = \frac{d}{dt} M \dot{\mathbf{R}} = 0
$$

Which means:

$$
    M \ddot{\mathbf{R}} = 0 \quad \Rightarrow \quad \ddot{\mathbf{R}} = 0
$$

Using the definition of the CM position, it follows that :

$$
    \ddot{\mathbf{R}} = \frac{d^2}{dt^2} \frac{m_1 \mathbf{r}_1 + m_2 \mathbf{r}_2}{M} = \frac{m_1 \ddot{\mathbf{r}}_1 + m_2 \ddot{\mathbf{r}}_2}{M}
$$

We can see that an equally valid equation follows:

$$
    \ddot{\mathbf{R}} = \frac{1}{M} \left( \mathbf{F}_{12} + \mathbf{F}_{21} \right)
$$

Defining the relative position vector as:

$$
    \mathbf{r} = \mathbf{r}_1 - \mathbf{r}_2
$$

We can then state that the two positions are as follows:

$$
    \mathbf{r}_1 = \mathbf{R} + \frac{m_2}{M} \mathbf{r} \\
    \mathbf{r}_2 = \mathbf{R} - \frac{m_1}{M} \mathbf{r}
$$

With two accelerations as:

$$
    \ddot{\mathbf{r}}_1 = \ddot{\mathbf{R}} + \frac{m_2}{M} \ddot{\mathbf{r}} \\
    \ddot{\mathbf{r}}_2 = \ddot{\mathbf{R}} - \frac{m_1}{M} \ddot{\mathbf{r}}
$$

We can show the following steps:

$$
    \tag{1}
    M \ddot{\mathbf{R}} = m_1 \ddot{\mathbf{r}}_1 + m_2 \ddot{\mathbf{r}}_2
$$

$$
    \tag{2}
    \ddot{\mathbf{R}} \left( m_1 + m_2 \right) = m_1 \ddot{\mathbf{r}}_1 + m_2 \ddot{\mathbf{r}}_2
$$

$$
    \tag{3}
    \ddot{\mathbf{R}} \left( m_1 + m_2 \right) = \mathbf{F}_{12} + m_2 \left( \ddot{\mathbf{R}} - \frac{m_1}{M} \ddot{\mathbf{r}} \right)
$$

$$
    \tag{4}
    \ddot{\mathbf{R}} \left( m_1 + m_2 \right) = \mathbf{F}_{12} + m_2  \ddot{\mathbf{R}} - \frac{m_1 m_2}{M} \ddot{\mathbf{r}}
$$

$$
    \tag{5}
    \mu = \frac{m_1 m_2}{M}
$$

$$
    \tag{6}
    \ddot{\mathbf{R}} m_1 = \mathbf{F}_{12} - \mu \ddot{\mathbf{r}}
$$

We know that $\ddot{\mathbf{R}} = 0$, which means:

$$
    \tag{7}
    0 = \mathbf{F}_{12} - \mu \ddot{\mathbf{r}}

$$

$$

$$

We are given the radial equation as:

$$
    \mu r \dot{\phi}^2 - \frac{dV}{dr} = \mu \ddot{r}
$$

We are also given the equation:

$$
    \dot{\phi} = \frac{\mathcal{l}}{\mu r^2}
$$

In Taylor, we are shown that we can define the centrifugal force as:

$$
    F_{cf} = \mu r \dot{\phi}^2 \equiv \frac{\mathcal{l}^2}{\mu r^3}
$$

We can, as well, define the centrifugal force in terms of a centrifugal potential energy:

$$
    F_{cf} = -\frac{d}{dr} \left( \frac{\mathcal{l}^2}{2 \mu r^2} \right) = -\frac{d V_{cf}}{dr}
$$

This means we can re-write the radial equation as:

$$
    \mu \ddot{r} = \frac{d}{dr} \left( V(r) + V_{cf} \right)
$$

Which means we can define the effective potential as:

$$
    V_{\text{eff}} = V(r) + V_{cf} = V(r) + \left( \frac{\mathcal{l}^2}{2 \mu r^2} \right)
$$

```python
# Exercise 4 Code

import matplotlib.pyplot as plt
import numpy as np

def V(k, r):
    return (0.5 * k * (r**2))


def V_cf(l, m, r):
    return ((l**2) / (m * (r**3)))


def V_eff(k, l, m, r):
    return V(k, r) + V_cf(l, m, r)


def calculate_velocity_verlet(dim=1, dt=0.01, tf=10, ti=0, r0=0, v0=0):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros((n, dim), dtype=float)
    v = np.zeros((n, dim), dtype=float)
    E = np.zeros((n, dim), dtype=float)
    
    # Before we calculate anything, we need to set up a few constants.
    k = 0.1
    m = 0.1
    l = 1
    omega = k/m
    C1 = l**2 / m**2
    C2 = l**2 / m
    r_min = (l**2 / (m*k))**0.25
    
    # Setting up the initial position and velocity vectors.
    r[0] = r_min
    v[0] = 0
    E[0] = V_eff(k, l, m, r[0])
    
    for i in range(n - 1):
        # a = get_acceleration(r[i])
        a = -r[i] * omega + C1 / (r[i]**3) 
        r[i+1] = r[i] + v[i]*dt + 0.5*a * (dt ** 2)
        
        ap = -r[i+1] * omega + C1 / (r[i+1]**3) 
        v[i+1] = v[i] + 0.5*(a + ap)*dt
        
        t[i+1] = t[i] + dt
        E[i+1] = V_eff(k, l, m, r[i]) + (0.5 * m * v[i+1]**2)
        
    return (t, r, v, E)

(t, r, v, E) = calculate_velocity_verlet(dim=1, dt=0.001, tf=10, ti=0)

plt.rcParams['figure.constrained_layout.use'] = True
(fig, ax) = plt.subplots(nrows=1, ncols=2)

ax[0].plot(t, r)
ax[0].set_xlabel("Time [s]")
ax[0].set_ylabel("Position [m]")
ax[0].set_title("Constant Radius vs Time")

ax[1].plot(t, E)
ax[1].set_xlabel("Time [s]")
ax[1].set_ylabel("Energy [J]")
ax[1].set_title("Energy vs Time")
plt.show()
```
