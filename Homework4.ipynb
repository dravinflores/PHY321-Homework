* Euler's Method
$$
    \mathbf{v} = \int_{t_0}^{t} \mathbf{a} \  dt = \int_{t_0}^{t} \frac{d\mathbf{v}}{dt} dt \approx \mathbf{v}\left(t_0\right) + \mathbf{a}(t) \Delta t

$$

$$

$$

* Velocity Verlet Method
$$
    \tag{1}
    x_{i+1} = x_i + v_{x, i} \Delta t + \frac{\Delta t^2}{2} a_{x, i}
$$

$$
    \tag{2}
    y_{i+1} = y_i + v_{y, i} \Delta t + \frac{\Delta t^2}{2} a_{y, i}
$$

$$
    \tag{3}
    v_{x, i+1} = v_i + \frac{\Delta t}{2}\left(a_{i+1} + a_i\right)
$$

$$
    \tag{4}
    v_{y, i+1} = v_i + \frac{\Delta t}{2}\left(a_{i+1} + a_i\right)
$$

```python
def calculate_euler(dim=2, dt=0.001, tf=10, ti=0, r0=0, v0=0):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # Just in-case we ever want to choose default values.
    if type(r0) is int:
        r0 = np.zeros((1, dim), dtype=float)
    if type(v0) is int:
        v0 = np.zeros((1, dim), dtype=float)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros((n, dim), dtype=float)
    v = np.zeros((n, dim), dtype=float)
    
    # This constant is used in the calculations below. Heavily. Thus, it is
    # easier to just define it now.
    k = 4 * math.pi * math.pi

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0

    for i in range(n - 1):
        r_mag = math.sqrt(sum(r[i]*r[i]))
        a = (-k * r[i]) / (r_mag ** 3)
        
        v[i+1] = v[i] + a*dt
        r[i+1] = r[i] + v[i]*dt
        t[i+1] = t[i] + dt
        
    return (t, r, v)


def calculate_euler_cromer(dim=2, dt=0.001, tf=10, ti=0, r0=0, v0=0):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # Just in-case we ever want to choose default values.
    if type(r0) is int:
        r0 = np.zeros((1, dim), dtype=float)
    if type(v0) is int:
        v0 = np.zeros((1, dim), dtype=float)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros((n, dim), dtype=float)
    v = np.zeros((n, dim), dtype=float)
    
    # This constant is used in the calculations below. Heavily. Thus, it is
    # easier to just define it now.
    k = 4 * math.pi * math.pi

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0

    for i in range(n - 1):
        r_mag = math.sqrt(sum(r[i]*r[i]))
        a = (-k * r[i]) / (r_mag ** 3)
        
        v[i+1] = v[i] + a*dt
        r[i+1] = r[i] + v[i+1]*dt
        t[i+1] = t[i] + dt
        
    return (t, r, v)


def calculate_velocity_verlet(dim=2, dt=0.001, tf=10, ti=0, r0=0, v0=0):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # Just in-case we ever want to choose default values.
    if type(r0) is int:
        r0 = np.zeros((1, dim), dtype=float)
    if type(v0) is int:
        v0 = np.zeros((1, dim), dtype=float)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros((n, dim), dtype=float)
    v = np.zeros((n, dim), dtype=float)
    
    # This constant is used in the calculations below. Heavily. Thus, it is
    # easier to just define it now.
    k = 4 * math.pi * math.pi

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        r_mag = math.sqrt(sum(r[i]*r[i]))
        a = (-k * r[i]) / (r_mag ** 3)
        
        r[i+1] = r[i] + v[i]*dt + 0.5*a * (dt ** 2)
        r_mag = math.sqrt(sum(r[i+1]*r[i+1]))
        ap = (-k * r[i+1]) / (r_mag ** 3)
        
        v[i+1] = v[i] + 0.5*(a + ap)*dt
        t[i+1] = t[i] + dt
        
    return (t, r, v)


# Intial Conditions
r0 = np.array([1.00, 0.00])
v0 = np.array([0.00, 2 * math.pi])

k = 8*math.pi*math.pi
v_esc = np.array([math.sqrt(k/r0[0]), math.sqrt(k/ (r0[0]) + 1 )])

(ta, ra, va) = calculate_euler(r0=r0, v0=v0)
(tb, rb, vb) = calculate_euler_cromer(r0=r0, v0=v0)
(tc, rc, vc) = calculate_velocity_verlet(r0=r0, v0=v0)

# Plot the data
# (figure, axis) = plt.subplots()
# axis.set_xlabel('x (m)')
# axis.set_ylabel('y (m)')
# axis.plot(ra[:, 0], ra[:, 1])
# figure.tight_layout()

plt.subplot(1, 3, 1)
plt.plot(ra[:, 0], ra[:, 1])
axis.set_xlabel('x (m)')
axis.set_ylabel('y (m)')
plt.title("Euler's Method")
figure.tight_layout()

plt.subplot(1, 3, 2)
plt.plot(rb[:, 0], rb[:, 1])
axis.set_xlabel('x (m)')
axis.set_ylabel('y (m)')
plt.title("Euler Cromer's Method")
figure.tight_layout()

plt.subplot(1, 3, 3)
plt.plot(rc[:, 0], rc[:, 1])
axis.set_xlabel('x (m)')
axis.set_ylabel('y (m)')
plt.title("Velocity Verlet's Method")
figure.tight_layout()

plt.show()
```
