* We are given the equation 
$$
    V(x) = \frac{V_0}{d^4} \left( x^4 - 2x^2 d^2 + d^4 \right)
$$
* The parameters $V_0$ and $d$ are known constants.
* Using the following Python code, the plot can be executed.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def V(V_0, d, x):
    return (V_0/(d**4)) * (x**4 - (2 * x**2 * d**2) + d**4)

V_0 = 1
d = 1

x = np.linspace(-2, 2, 500)
V = V(V_0, d, x)

plt.plot(x, V)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.ylim(-0, 5)
plt.show()
```

* If the particle starts at rest at $x = 2d$, then this means that the potential is:

$$
    V\left(2d\right) = \frac{V_0}{d^4} \left( \left(2d\right)^4 - 2 \left(2d\right)^2 d^2 + d^4 \right) = \frac{V_0}{d^4} \left( 16 d^4 - 8 d^4 + d^4 \right) = 9 \ V_0
$$

* We can then figure out the potential at the point $x = d$:

$$
    V(d) = 0
$$

* This means that at the point $x = d$, all the energy is kinetic energy. Thus:

$$
    E = K(v) = \frac{m v^2}{2} \quad \Rightarrow \quad v = \pm \sqrt{ \frac{2}{m} E_0 }
$$

4.

* If the particle starts at rest at the point $x = d$ with an initial velocity $v_0$, then to reach the point $x = -d$, the velocity must be large. This is due to the fact that $x = d$ is an equilibrium point. In order for the particle to reach the other well, the initial velocity must satisfy the following:

$$
    v = \pm \sqrt{ \frac{2}{m} E_0 }
$$

5.

---

* Given that we are extrapolating forces from a potential function, and that the potential function only depends upon the position, we know that this force is conservative. We can also use the following identity:

$$
    \mathbf{F} = -\nabla V; \quad \nabla \times \mathbf{F} = \nabla \times \left(-\nabla V\right) = 0 
$$

* The force can be found as:

$$
    \mathbf{F} = - \frac{d}{dx} \left( \frac{V_0}{d^4} \left( x^4 - 2x^2 d^2 + d^4 \right) \right) = - \frac{V_0}{d^4} \left( 4x^3 - 4x d^2 \right)
$$

* This means that the acceleration is:

$$
    \mathbf{a} = \frac{\mathbf{F}}{m} = - \frac{V_0}{m \ d^4} \left( 4x^3 - 4x d^2 \right)
$$

```python
# Euler-Cromer Method and the Velocity Verlet Method

import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
import math

%matplotlib inline

def V(V_0, d, x):
    try:
        return (V_0/(d**4)) * (x**4 - (2 * x**2 * d**2) + d**4)
    except:
        return -5


def get_acceleration(V_0, d, m, x, alt=False):
    if alt:
        return (-(V_0 / (m * d**4)) * (-4*x * d**2))
    else:
        return (-(V_0 / (m * d**4)) * (4 * x**3 - 4*x * d**2))


def calculate_euler_cromer(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0, V_0=1, d=1, m=1, alt=False):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros(n)
    v = np.zeros(n)

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        if not alt:
            a = get_acceleration(V_0, d, m, r[i])
        else:
            a = get_acceleration(V_0, d, m, r[i], True)
        
        v[i+1] = v[i] + a*dt
        r[i+1] = r[i] + v[i+1]*dt
        t[i+1] = t[i] + dt
        
    return (t, r, v)


def calculate_velocity_verlet(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0, V_0=1, d=1, m=1, alt=False):
    E_0 = (1/2)*(m)*(v0**2) + V(V_0, d, r0)
    
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros(n)
    v = np.zeros(n)
    E = np.zeros(n)

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        if not alt:
            a = get_acceleration(V_0, d, m, r[i])
        else:
            a = get_acceleration(V_0, d, m, r[i], True)
        
        r[i+1] = r[i] + v[i]*dt + 0.5*a * (dt ** 2)
        
        if not alt:
            ap = get_acceleration(V_0, d, m, r[i+1])
        else:
            ap = get_acceleration(V_0, d, m, r[i+1], True)
        
        v[i+1] = v[i] + 0.5*(a + ap)*dt
        t[i+1] = t[i] + dt
        E[i+1] = E_0 + (1/2)*(m)*(v[i]**2) + V(V_0, d, r[i])
        
    return (t, r, v, E)


"""
Implement the numerical algorithms in a program to find the position of the particle as function of time from t=0 to t=30 s 
using a mass m=1.0 kg, the parameter V0=1 J and d=0.1 m. Make a plot of three distinct positions with initial conditions x0=d 
and v0=0.5 m/s, x0=d and v0=1.5 m/s, and x0=d and v0=2.5 m/s. Plot also the velocity. Perform calculations with and without 
the term x4 in the potential. Do you see a difference? Compare and discuss the results obtained with the Velocity Verlet 
algorithm and the Euler-Cromer algorithm.
"""

tf = 30
ti = 0
m = 1
V0 = 1
d = 0.1

# These lists store the numpy arrays for both the Euler-Cromer and Velocity-Verlet methods.
te_list = list()
tv_list = list()
re_list = list()
rv_list = list()
ve_list = list()
vv_list = list()
Ev_list = list()

init_conditions = [(d, 0.5), (d, 1.5), (d, 2.5)]

for (x_0, v_0) in init_conditions:
    (te, re, ve) = calculate_euler_cromer(tf=30.0, ti=0.0, dt=0.001, r0=x_0, v0=v_0, V_0=V0, d=d, m=m, alt=False)
    (tv, rv, vv, Ev) = calculate_velocity_verlet(tf=30.0, ti=0.0, dt=0.001, r0=x_0, v0=v_0, V_0=V0, d=d, m=m, alt=False)
    
    te_list.append(te)
    tv_list.append(tv)
    re_list.append(re)
    rv_list.append(rv)
    ve_list.append(ve)
    vv_list.append(vv)
    Ev_list.append(Ev)
    
# Plot the stuff
matplotlib.style.use('seaborn')
plt.rcParams["figure.figsize"] = (20, 20)
(fig, axes) = plt.subplots(2, 2)

axes[0][0].plot(te, re)
axes[0][0].set_title('Euler-Cromer Position')

axes[0][1].plot(te, ve)
axes[0][1].set_title('Euler-Cromer Velocity')

axes[1][0].plot(tv, rv)
axes[1][0].set_title('Velocity-Verlet Position')

axes[1][1].plot(tv, vv)
axes[1][1].set_title('Velocity-Verlet Velocity')

plt.tight_layout()

init_conditions = [(d, 0.5), (d, 1.5), (d, 2.5)]

for (x_0, v_0) in init_conditions:
    (te, re, ve) = calculate_euler_cromer(tf=30.0, ti=0.0, dt=0.001, r0=x_0, v0=v_0, V_0=V0, d=d, m=m, alt=True)
    (tv, rv, vv, Ev) = calculate_velocity_verlet(tf=30.0, ti=0.0, dt=0.001, r0=x_0, v0=v_0, V_0=V0, d=d, m=m, alt=True)
    
    te_list.append(te)
    tv_list.append(tv)
    re_list.append(re)
    rv_list.append(rv)
    ve_list.append(ve)
    vv_list.append(vv)
    Ev_list.append(Ev)

    

# Plot the stuff. 
matplotlib.style.use('seaborn')
plt.rcParams["figure.figsize"] = (15, 15)
(fig, axes) = plt.subplots(2, 2)

axes[0][0].plot(te_list[5], re_list[5])
axes[0][0].set_title('Euler-Cromer Position')

axes[0][1].plot(te_list[5], ve_list[5])
axes[0][1].set_title('Euler-Cromer Velocity')

axes[1][0].plot(tv_list[5], rv_list[5])
axes[1][0].set_title('Velocity-Verlet Position')

axes[1][1].plot(tv_list[5], vv_list[5])
axes[1][1].set_title('Velocity-Verlet Velocity')

plt.tight_layout()
plt.show()
```

## Part 3

* We have the potential given as:

$$
    V(x) = \frac{V_0}{d^4} \left( x^4 - 2x^2 d^2 + d^4 \right)
$$

* The force can be found as:

$$
    \mathbf{F} = -\frac{dV}{dt} = \frac{V_0}{d^4} \left( 4x^3 - 4x d^2 \right)
$$

* The acceleration can be found as 

$$
    \mathbf{a} = \frac{\mathbf{F}}{m} = \frac{V_0}{m d^4} \left( 4x^3 - 4x d^2 \right)
$$

* We can add up the internal forces and equate them to the driven force as:

$$
    \frac{V_0}{d^4} \left( 4x^3 - 4x d^2 \right) + \beta v = F_0 \cos{\left(\omega t\right)}
$$

* We can then set up the differential equation as:

$$
    \mathbf{F} = F_0 \cos{\left(\omega t\right)} - \beta v - \frac{V_0}{d^4} \left( 4x^3 - 4x d^2 \right)
$$

```python
# Part 3 Code which implements the Runge-Kutta Method.

import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
import math

%matplotlib inline

matplotlib.style.use('seaborn')
plt.rcParams["figure.figsize"] = (15, 15)

def V(V_0, d, x):
    try:
        return (V_0/(d**4)) * (x**4 - (2 * x**2 * d**2) + d**4)
    except:
        return -5
    
def F(F_0, d, omega, beta, v, x, t):
    # ma + bv + kr = F_0 cos(wt)
    return ((F_0 * math.cos(omega * t)) - (beta*v) - ((V_0)/(d**4) * (4 * x**3 - 4 * x * d**2)))
    
    
def get_acceleration(F_0, d, omega, beta, m, v, x, t):
    return (F(F_0, d, omega, beta, v, x, t) / m)


def calculate_euler_cromer(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0, V_0=1, d=1, m=1, F_0=1, omega=1, beta=0.1):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros(n)
    v = np.zeros(n)

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        a = get_acceleration(F_0, d, omega, beta, m, v[i], r[i], t[i])
        
        v[i+1] = v[i] + a*dt
        r[i+1] = r[i] + v[i+1]*dt
        t[i+1] = t[i] + dt
        
    return (t, r, v)


def RK4(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0, V_0=1, d=1, m=1, F_0=1, omega=1, beta=0.1):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    x = np.zeros(n)
    v = np.zeros(n)

    # Setting up the initial position and velocity vectors.
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        # The calculation for K1
        K1x = v[i]*dt
        K1v = get_acceleration(F_0, d, omega, beta, m, v[i], x[i], t[i]) * dt
        
        # Calculation for K2
        xi = x[i] + K1x*0.5
        vi = v[i] + K1v*0.5
        K2x = vi*dt
        K2v = get_acceleration(F_0, d, omega, beta, m, vi, xi, t[i] + 0.5*dt) * dt
        
        # Calculation for K3
        xi = v[i] + K2x*0.5
        vi = v[i] + K2v*0.5
        K3x = vi*dt
        K3v = get_acceleration(F_0, d, omega, beta, m, vi, xi, t[i] + 0.5*dt) * dt
    
        # Calculation for K4
        xi = x[i] + K3x
        vi = v[i] + K3v
        K4x = vi*dt
        K4v = get_acceleration(F_0, d, omega, beta, m, vi, xi, t[i]+dt) * dt
        

        x[i+1] = x[i] + ((K1x + 2*K2x + 2*K3x + K4x) / 6)
        v[i+1] = v[i] + ((K1v + 2*K2v + 2*K3v + K4v) / 6)
        t[i+1] = t[i] + dt
        
    return (t, x, v)


initial_conditions = [(0.1, 0.5), (0.1, 1.5), (0.1, 2.5)]
F_list = [1, -1, 0.000001]
omega_list = [0, 0, -math.pi]
t_list = list()
x_list = list()
v_list = list()

for i in range(3):
    (t, x, v) = RK4(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0, V_0=1, d=1, m=1, F_0=F_list[i], omega=omega_list[i], beta=0.1)
    t_list.append(t)
    x_list.append(x)
    v_list.append(v)

(fig, ax) = plt.subplots(6, constrained_layout=True)

for i in range(3):
    (t, x, v) = calculate_euler_cromer(tf=10.0, ti=0.0, dt=0.001, r0=0, v0=0, V_0=1, d=1, m=1, F_0=F_list[i], omega=omega_list[i], beta=0.1)
    t_list.append(t)
    x_list.append(x)
    v_list.append(v)

title_list = [
    "Runge-Kutta: Position vs. Time, stays in right well.", 
    "Runge-Kutta: Position vs. Time, stays in left well.",
    "Runge-Kutta: Position vs. Time, jumps wells.",
    "Euler-Cromer: Position vs. Time, stays in right well.", 
    "Euler-Cromer: Position vs. Time, stays in left well.",
    "Euler-Cromer: Position vs. Time, jumps wells."
]

for i in range(3):
    ax[i].plot(t_list[i], x_list[i])
    title = title_list[i]
    ax[i].set_title(title)
    
for i in range(3, 6):
    ax[i].plot(t_list[i], x_list[i])
    title = title_list[i]
    ax[i].set_title(title)
    
plt.show()
```

* We can produce a differential equation:

$$
    \frac{d^2 x}{dt^2} + \gamma \frac{dx}{dt} = \alpha x + \beta x^2 + \delta x^3
$$

* To get an understanding of what a solution might look like, we will use the Euler-Cromer method to numerically solve the differential equation. The acceleration can be found as:

$$
    \frac{d^2 x}{dt^2} = \left( \alpha x + \beta x^2 + \delta x^3 \right) - \gamma \frac{dx}{dt}
$$

    x(t) = x_h(t) + x_p(t)

* We know that the homogeneous solution $x_h(t)$ can be found as follows:

$$
    \tag{1}
    \frac{d^2 x}{dt^2} + \gamma \frac{dx}{dt} = 0
$$

$$
    \tag{2}

$$

$$
    \tag{3}
    p(r) = 0, \ r = \{ \ -\gamma, \ 0 \ \}
$$

$$
    \tag{4}
    x_h(t) = C_1 e^{0} + C_2 e^{-\gamma t} = C_1 + C_2 e^{-\gamma t}
$$

* So far, we have:
$$
    x(t) = C_1 + C_2 e^{-\gamma t} + x_p(t)
$$

* The particular solution requires a bit more effort. We know that the particular solution should have a similar form to:

$$
    \alpha x + \beta x^2 + \delta x^3
$$

* Given that $\beta = 0$, we can alternatively say that $x_p(t)$ should have a form similar to:

$$
    \alpha x + \delta x^3 = \alpha x(t) + \delta \left( x(t) \right)^3
$$

* This would give us the form:

$$
    x_p(t) = K_1 x(t) + K_2 \left(x(t)\right)^3
$$

* This form does not look like a solvable equation. Back to the drawing board.

* We have the equation:

$$
    \frac{d^2 x}{dt^2} + \gamma \frac{dx}{dt} = \alpha x + \beta x^2 + \delta x^3
$$

* Which can be algebraically manipulated to say:

$$
    \frac{d^2 x}{dt^2} = \left( \alpha x + \beta x^2 + \delta x^3 \right) - \gamma \frac{dx}{dt}
$$

* Which we can then say:

$$
    \frac{d^2 x}{dt^2} = f\left( x, \frac{dx}{dt} \right)
$$

* In this manner, we can decouple this function using the following relations:

$$
    \frac{dx}{dt} = v, \quad \frac{dv}{dt} = f(x, v)
$$

* Which we can then state as:

$$
    \frac{dx}{dt} = v, \quad \frac{dv}{dt} =  \left( \alpha x + \delta x^3 \right) - \gamma v
$$

* Before we continue, we should do some dimensional analysis on the constants, to confirm that our statement is valid:

$$
    \left[ \frac{dv}{dt} \right] = \text{m/}\text{s}^2 \quad \Rightarrow \quad \left[ \alpha \right] = \text{s}^{-2}, \ \left[ \delta \right] = \text{m}^{-2}\text{s}^{-2}, \ \left[ \gamma \right] = \text{s}^{-1}
$$
