## Exercise 1
* We introduce the additional requirement, for a force to be conservative, the curl of that force must be zero. That is 

$$
    \nabla \times \mathbf{F} = 0
$$

---

1A.

---

* $\mathbf{F} = k \cdot \left( x, y, z \right) = \left( kx, ky, kz \right)$ 

$$
    \nabla \times \mathbf{F} = \left(\frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z}\right) \times \left( kx, ky, kz \right) = \\ 
    \left(\frac{\partial}{\partial y} kz - \frac{\partial}{\partial z} ky \right) \hat{\mathbf{i}} - \left( \frac{\partial}{\partial x} kz - \frac{\partial}{\partial z} kx \right) \hat{\mathbf{j}} + \left( \frac{\partial}{\partial x} ky - \frac{\partial}{\partial y} kx \right) \hat{\mathbf{k}} = \mathbf{0}
$$

* This force is conservative as the curl of this force is equivalent to the zero vector.


---

1B.

---

* $\mathbf{F} = \left( Ax, By^2, Cz^2 \right)$ 


$$
    \nabla \times \mathbf{F} = \left(\frac{\partial}{\partial y} Cz^2 - \frac{\partial}{\partial z} By^2 \right) \hat{\mathbf{i}} - \left( \frac{\partial}{\partial x} Cz^2 - \frac{\partial}{\partial z} Ax \right) \hat{\mathbf{j}} + \left( \frac{\partial}{\partial x} By^2 - \frac{\partial}{\partial y} Ax \right) \hat{\mathbf{k}} = \mathbf{0}
$$

* This force is also conserative as the curl of this force is equivalent to the zero vector.

---

1C.

---

* $\mathbf{F} = \left( Ay, Bx, Cz \right)$ 

$$
    \nabla \times \mathbf{F} = \left(\frac{\partial}{\partial y} Cz - \frac{\partial}{\partial z} Bx \right) \hat{\mathbf{i}} - \left( \frac{\partial}{\partial x} Cz - \frac{\partial}{\partial z} Ay \right) \hat{\mathbf{j}} + \left( \frac{\partial}{\partial x} Bx - \frac{\partial}{\partial y} Ay \right) \hat{\mathbf{k}} = 0 \ \hat{\mathbf{i}} + 0 \ \hat{\mathbf{j}} + \left( B - A \right) \ \hat{\mathbf{k}}
$$

* This force will only be conservative if $A = B$. Otherwise, the curl of this force is not equivalent to the zero vector, and therefore is not a conservative force.

$$

* This force is conservative.
