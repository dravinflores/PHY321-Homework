$$

    \left(\frac{\partial}{\partial y} kz - \frac{\partial}{\partial z} ky \right) \hat{\mathbf{i}} - \left( \frac{\partial}{\partial x} kz - \frac{\partial}{\partial z} kx \right) \hat{\mathbf{j}} + \left( \frac{\partial}{\partial x} ky - \frac{\partial}{\partial y} kx \right) \hat{\mathbf{k}} = \mathbf{0}

$$

$$

* In this case, we will take a step toward scaling the equation. We have:

$$
    \frac{d^2 x}{dt^2} + \omega_0^2 x = f_0 cos(\omega t)
$$

* I will first introduce the dimensionless time:

$$
    \frac{d^2 x}{d\tau^2} + x = \frac{f_0}{\omega_0^2} cos(\omega \tau)
$$

* And now the dimensionless frequency:

$$
    \frac{d^2 x}{d\tau^2} + x = \tilde{F} cos(\tilde{\omega} \tau)
$$

```python
# Exercise 5 Code

import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

def get_potential(alpha, beta, x):
    return (-alpha * (x**2 / 2)) + (beta * (x**4 / 4))

def get_kinetic(m, v):
    return (1/2)* m * v**2

def get_force_original(alpha, beta, x):
    return alpha*x - (beta * x**3)

# The force where we just ignore the higher powers of x.
def get_force(alpha, x):
    return alpha*x

def get_analytic(f_0, omega_0, omega, t):
    A = f_0 / (omega_0**2 - omega^2)**2
    return A * math.cos(omega * t) - A * math.sin( (math.sqrt(2)/2) * omega_0**2 * t)


def calculate_euler_cromer():
    tf = 20
    ti = 0
    dt = 0.01
    
    F_0 = 1
    m = 1
    
    f_0 = 2
    b = 0
    k = 1
    omega = 2
    omega_0 = 25
    omega_0_small = 1
    
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    x = np.zeros(n)
    v = np.zeros(n)
    v_o = np.zeros(n)
    x_a = np.zeros(n)
    x_o = np.zeros(n)

    # Setting up the initial position and velocity vectors.
    x[0] = 0
    v[0] = 0
    
    for i in range(n - 1):
        a = f_0 * math.cos(omega * t[i]) - (omega_0**2 * x[i])
        a_o = f_0 * math.cos(omega * t[i]) - (omega_0_small**2 * x[i])
        
        x_a[i] = get_analytic(f_0, omega_0, omega, t[i]) 
        
        v[i+1] = v[i] + a*dt
        x[i+1] = x[i] + v[i+1]*dt
        
        v_o[i+1] = v_o[i] + a_o*dt
        x_o[i+1] = x_o[i] + v_o[i+1]*dt
        
        t[i+1] = t[i] + dt
        
    return (t, x, v, x_a, x_o)

(t, x, v, x_a, x_o) = calculate_euler_cromer()

'''
(fig, axis) = plt.subplots()
axis.set_ylabel('x (m)')
axis.set_xlabel('t (s)')
axis.plot(t, x)
fig.tight_layout()
plt.show()
'''

plt.subplot(1, 3, 1)
plt.plot(t, x_a)
plt.xlabel('t (s)')
plt.ylabel('x (m)')
plt.title('Analytic')

plt.subplot(1, 3, 2)
plt.plot(t, x)
plt.xlabel('t (s)')
plt.ylabel('x (m)')
plt.title('Numerical')

plt.subplot(1, 3, 3)
plt.plot(t, x_o)
plt.xlabel('t (s)')
plt.ylabel('x (m)')
plt.title('Numerical Odd')

plt.show()
```
