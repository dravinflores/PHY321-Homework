We have

$$
    \sqrt{ \left(x - a\right)^2 + y^2} + \sqrt{ \left(x + a\right)^2 + y^2} = 2D
$$

Which can be re-written to be

$$
    \sqrt{ \left(x + a\right)^2 + y^2} = 2D - \sqrt{ \left(x - a\right)^2 + y^2}
$$

We can then square both sides to obtain

$$
    \left(x + a\right)^2 + y^2 = a^2 + 4 D^2 - 2 a x + x^2 + y^2 - 4D \sqrt{(x - a)^2 + y^2}
$$

Which can be expanded to be

$$
    a^2 + 2 a x + x^2 + y^2 = a^2 + 4 D^2 - 2 a x + x^2 + y^2 - 4D \sqrt{(x - a)^2 + y^2}
$$

We can then simplify and cancel terms to obtain

$$
    D \sqrt{a^2 - 2 a x + x^2 + y^2} = D^2 - a x
$$

We can square both sides to obtain

$$
    a^2 D^2 - 2 a D^2 x + D^2 x^2 + D^2 y^2 = D^4 - 2 a D^2 x + a^2 x^2
$$

Which can be then written as

$$
    D^2 x^2 + D^2 y^2 - a^2 x^2 = D^4 - a^2 D^2
$$

Which we can simplify to

$$
   x^2 \left( D^2 - a^2 \right) + D^2 y^2 = D^2 \left( D^2 - a^2 \right)
$$

We can then divide the left side by the right side to produce

$$
    \frac{x^2}{D^2} + \frac{y^2}{\left( D^2 - a^2 \right)} = 1
$$

## Exercise 3

> ### A.

We are given the equation of motion as

$$
    \ddot{r} = - \frac{1}{m} \frac{dV}{dr} + r \dot{\phi}^2
$$

Which can then be multiplied by mass to get the equation for the net force

$$
    m \ddot{r} = -\frac{dV}{dr} + m r \dot{\phi}^2
$$

We can recognize this as

$$
    F = -\frac{dV}{dr} + F_{\text{cf}}
$$

We are given the equation for $\dot{\phi}$ as

$$
    \dot{\phi} = \frac{L}{m r^2}
$$

This means that we can find $\dot{\phi}^2$ as

$$
    \dot{\phi}^2 = \frac{L}{m r^2} \cdot \frac{L}{m r^2} = \frac{L^2}{m^2 r^4}
$$

We can then state that the centrifugal force is

$$
    F_{\text{cf}} = m r \dot{\phi}^2 = \frac{L^2}{m r^3}
$$

Which we can alternatively state that this force is the derivative of the centrifugal potential with respect to radius, which is

$$
    F_{\text{cf}} = - \frac{d}{dr} \left( \frac{L^2}{2 m r^2} \right)
$$

In this manner, we have the net force as

$$
    F = - \frac{d}{dr} V + -\frac{d}{dr} V_{\text{cf}}
$$

Which means we can define an effective potential as

$$
    V_{\text{eff}} = V + V_{\text{cf}} = -\frac{\alpha}{r} + \frac{L^2}{2 m r^2}
$$

The position of the minimum effective potential means that

$$
    -\frac{d}{dr} V_{\text{eff}} = 0
$$

Which means that

$$
    -\frac{d}{dr} V + -\frac{d}{dr} V_{\text{cf}} = 0
$$

Which requires

$$
    \frac{\alpha}{r^2} = \frac{L^2}{m r^3}
$$

Which can be algebraically manipulated to state that

$$
    r = \frac{L^2}{m \alpha}
$$

> ### B.

As we know that $\ddot{r} = 0$ at the position stated above, we can then begin to solve the equation

$$
    \ddot{r} = 0 = \frac{F}{m} + \dot{\phi}^2 r
$$

Which is equivalently

$$
    -\frac{\alpha}{m r^2} + \dot{\phi}^2 r = 0
$$

As the radial velocity is zero, this means that we can rearrange this equation to state

$$
    \dot{\phi} = \sqrt{\frac{\alpha}{m r^3}}
$$

We can plug in the radius found in A. to obtain

$$
    \dot{\phi} = \sqrt{\frac{\alpha}{m \left({\frac{L^2}{m \alpha}}^3\right)}}
$$

Which can be simplified to

$$
    \dot{\phi} = \frac{\alpha^2 m}{L^3}
$$

> ### C.

The minimum radius is given as

$$
    r_{\text{min}} = \frac{L^2}{m \alpha}
$$

Plugging in this minimum radius into the effective potential, we have

$$

$$

In this case, we can assert that the spring constant is $\alpha$, which is given by:

$$
    \alpha = \sqrt{ V_{\text{eff}} \left( r_{\text{min}} \right) \frac{2 L^2}{m} }

```python
# Code for Exercise 6

import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline

def V(alpha, r):
    return -(alpha / r)


def V_cf(L, m, r):
    return (L**2) / (2 * m * r**2)


def V_eff(alpha, L, m, r):
    return V(alpha, r) + V_cf(L, m, r)


def get_acceleration(alpha, L, m, r):
    return (-alpha / r**2) + (L**2 / (m**2 * r**3))


def calculate_velocity_verlet(dim=1, dt=0.01, tf=100, ti=0, r0=0, v0=0):
    # We need to know how many intervals to calculate.
    n = math.ceil( (tf - ti) / dt)

    # These are the arrays that will store our data.
    t = np.zeros(n)
    r = np.zeros((n, dim), dtype=float)
    v = np.zeros((n, dim), dtype=float)
    E = np.zeros((n, dim), dtype=float)
    
    # Before we calculate anything, we need to set up a few constants.
    L = 1
    m = 1
    a = 10
    
    # Gather the initial conditions
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        a = get_acceleration(a, L, m, r[i])
        r[i+1] = r[i] + v[i]*dt + 0.5*a * (dt ** 2)
        
        ap = get_acceleration(a, L, m, r[i+1])
        v[i+1] = v[i] + 0.5*(a + ap)*dt
        
        t[i+1] = t[i] + dt
        E[i+1] = V_eff(a, L, m, r[i]) # + (0.5 * m * v[i]**2)
    
        
    return (t, r, v, E)

# (t, r, v, E) = calculate_velocity_verlet(dim=1, dt=0.001, tf=10, ti=0, r0=1, v0=0)
# plt.rcParams['figure.constrained_layout.use'] = True
# (fig, ax) = plt.subplots(nrows=1, ncols=3)

# ax[0].plot(r, V_eff(0.5, 1, 1, r))
# ax[0].set_xlabel("Position")
# ax[0].set_ylabel("Potential")
# ax[0].set_title("Potential vs Position")

# ax[1].plot(t, r)
# ax[1].set_xlabel("Time")
# ax[1].set_ylabel("Position")
# ax[1].set_title("Radius vs Time")

# ax[2].plot(t, v)
# ax[2].set_xlabel("Time")
# ax[2].set_ylabel("Velocity")
# ax[2].set_title("Velocity vs Time")
# plt.show()
```

The eccentricity constant $\epsilon$ gives us an idea of the shape of the ellipse. We also know that the eccentricity is a characteristic of the energy.

We have four cases:

$$
    \epsilon = 0, \\
    0 < \epsilon < 1, \\
    \epsilon = 1, \\
    \epsilon > 1
$$

In each of the four cases, Taylor summarizes the result as follows

| Eccentricity       | Energy   | Orbit     |
|--------------------|----------|-----------|
| $\epsilon = 0$     | $E < 0$  | Circle    |
| $0 < \epsilon < 1$ | $E < 0$  | Ellipse   |
| $\epsilon = 1$     | $E = 0$  | Parabola  |
| $\epsilon > 1$     | $E > 0$  | Hyperbola |


### Elliptical Orbit

* This occurs when $0 < \epsilon < 1$
* When this occurs, we can see that $E < 0$ from the Energy equation in 4A.
* When $E < 0$, we know that the Energy of the system is constant and the radius of the particle will oscillate between two radii: $r_{\text{min}}$ and $r_{\text{max}}$
* In this case, we know that these two radii can be found using the formula

$$
    r_{\text{min}} = \frac{c}{1 + \epsilon}, \ r_{\text{max}} = \frac{c}{1 - \epsilon}
$$

* As we are in polar coordinates, when converting to cartesian coordinates, we see the elliptical pattern
* In the plots above, it appears that the elliptical orbit has a constantly changing position versus time.
* This means that the velocity is constant and the acceleration is zero. This is what we would expect for an object that is bound.

### Parabola
* This occurs when $\epsilon = 1$
* We see that $E = 0$
* In this case, there is a minimum radius. However, the maximum radius tends toward infinity. In this manner, we see that the orbit is unbounded.
* When converting from polar to cartesian, we find that the equation for the position is

$$
    y^2 = c^2 - 2cx
$$

* In the plots above, it appears that the rate by which the slope of the position graph changes is itself changing.
* This means that there is an acceleration. This is what we would expect.

### Hyperbola
* This occurs when $\epsilon > 1$
* We see that $E > 0$
* This causes the radial equation to require that the angle $\phi$ is bounded
* When converting from polar to cartesian, we find that the equation for the position is given as

$$
    \frac{\left(x - \delta\right)^2}{\alpha^2} - \frac{y^2}{\beta^2} = 1
$$

* It appears that in the plots above, the rate by which the position is changing is also changing.
* We can see that the acceleration of the object is constant.
